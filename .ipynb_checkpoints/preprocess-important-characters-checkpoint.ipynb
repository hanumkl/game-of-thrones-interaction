# Important Characters in Game of Thrones

```python
import pandas as pd
import numpy as np
```

```python
file_name = 'asoiaf-book{}-edges.csv'
df_raw = pd.concat([pd.read_csv(file_name.format(i)) for i in range(1, 6)])
df_raw.replace('-',' ',regex=True,inplace=True)
df_raw.rename(columns={'weight': 'Weight', 'book': 'Book', 'Source':'Character Name'}, inplace=True)
df_raw.head()
```

```python
df_raw.head()
df_raw.tail()
```

```python
df_add = df_raw.copy()
df_add.rename(columns={'Character Name': 'Target', 'Target': 'Character Name'}, inplace=True)
df_add.head()
```

```python
df = pd.concat([df_raw, df_add], sort='False')
df.head()
```

```python
df.info()
```

```python
df.tail()
```

```python
df_targaryen = df[df['Target'].str.contains(r'^(?=.*Targaryen)')]
df_targaryen['Family'] = 'Targaryen'

df_stark = df[df['Target'].str.contains(r'^(?=.*Stark)')]
df_stark['Family'] = 'Stark'

df_baratheon = df[df['Target'].str.contains(r'^(?=.*Baratheon)')]
df_baratheon['Family'] = 'Baratheon'

df_lannister = df[df['Target'].str.contains(r'^(?=.*Lannister)')]
df_lannister['Family'] = 'Lannister'

df_tyrell = df[df['Target'].str.contains(r'^(?=.*Tyrell)')]
df_tyrell['Family'] = 'Tyrell'

df_martell = df[df['Target'].str.contains(r'^(?=.*Martell)')]
df_martell['Family'] = 'Martell'

df_all_families= pd.concat([df_targaryen, df_stark, df_baratheon, df_lannister])
df_all_families
```

```python
df_all_families.info()
```

```python
agg_table = pd.pivot_table(df_all_families, values='Weight', index=['Book','Character Name'], 
            columns='Family', aggfunc=np.sum, fill_value=0)
agg_table.head()
```

```python
agg_table.info()
```
